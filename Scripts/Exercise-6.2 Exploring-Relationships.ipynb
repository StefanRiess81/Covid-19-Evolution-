# 6. Create a categorical plot and interpret the results.

```python
# Use a histogram to visualize the distribution of the dependent variable - total vaccinations

sns.histplot(df['total_vaccinations'], bins = 5, kde = True)
```

```python
#Creating a categorical variable that splits the "total_vaccinations" column into categories

df.loc[df['total_vaccinations'] < 10000, 'Vaccination Rate'] = 'Low Rate'
df.loc[(df['total_vaccinations'] >= 10000) & (df['total_vaccinations'] < 50000), 'Vaccination Rate'] = 'Middle Rate'
df.loc[df['total_vaccinations'] >= 50000, 'Vaccination Rate'] = 'High Rate'
```

```python
df['Vaccination Rate'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the total_vaccinations and other variables

sns.set(style="ticks")
sns.scatterplot(x = 'total_vaccinations', y = 'total_boosters', hue = 'Vaccination Rate', data = df)
```
